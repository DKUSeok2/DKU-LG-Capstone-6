```python
import pandas as pd

def preprocess_and_calculate_miscarriage(data):
    # 문자열에서 숫자만 추출 (예: "0회" → 0, "1회" → 1)
    def extract_number(value):
        if "이상" in value:
            return 6  # "6회 이상"은 6으로 처리
        return int(value.replace("회", ""))

    # 복사본 생성(본인이 불러온 dataset)
    df = data.copy()

    # 숫자 값으로 변환
    df['총 시술 횟수'] = df['총 시술 횟수'].apply(extract_number)
    df['총 임신 횟수'] = df['총 임신 횟수'].apply(extract_number)
    df['총 출산 횟수'] = df['총 출산 횟수'].apply(extract_number)

    # 유산 여부 계산
    df['유산 여부'] = df['총 임신 횟수'] > df['총 출산 횟수']
    # df['유산 횟수'] = df['총 임신 횟수'] - df['총 출산 횟수']

    # 불임 여부 계산
    df['불임 여부'] = df['총 시술 횟수'] > df['총 임신 횟수']
    # df['불임 횟수'] = df['총 시술 횟수'] - df['총 임신 횟수']


    return df
```

```python
data = preprocess_and_calculate_miscarriage(df)
data
```
